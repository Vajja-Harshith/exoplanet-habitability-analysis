```python
import pandas as pd

# Load the uploaded NASA exoplanet dataset
# Note: If your file is in a folder or has a slightly different path,
# Colab will let you right-click the file and choose "Copy path"
file_path = 'dataexoplanets.csv'

# Read the CSV, skipping any introductory NASA header comment lines (usually marked with #)
try:
    df = pd.read_csv(file_path, comment='#')
    print("Dataset loaded successfully")
    print(f"Total planets in dataset: {df.shape[0]}")
    print("\nfirst few columns available in dataset:")
    print(df.columns.tolist()[:30]) # Showcases the first 30 columns
except Exception as e:
    print("Error loading file. Make sure the file name matches exactly on the left sidebar!")
    print(e)
```

```python
# Filter out rows where the planet radius (pl_rade) is missing
clean_df = df.dropna(subset=['pl_rade']).copy()

# Filter for planets that are terrestrial/rocky (roughly between 0.5 and 1.5 times Earth's radius)
#make sure we are looking at unique planets (dropping duplicates if the same planet was observed multiple times)
rocky_planets = clean_df[(clean_df['pl_rade'] >= 0.5) & (clean_df['pl_rade'] <= 1.5)].drop_duplicates(subset=['pl_name'])

print(f"found {len(rocky_planets)} unique planets that are a similar size to Earth!")
print("\n10 top candidates with their discovery years and methods:")
print(rocky_planets[['pl_name', 'hostname', 'pl_rade', 'disc_year', 'discoverymethod']].head(10).to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plotting style
plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Create a scatter plot of all your Earth-sized worlds
# We will use 'pl_orbper' (orbital period in days) on the X-axis
# and 'pl_rade' (planetary radius) on the Y-axis
sns.scatterplot(
    data=rocky_planets,
    x='pl_orbper',
    y='pl_rade',
    alpha=0.6,
    color='blue'
)

# Since some planets have massive orbital periods, we will restrict the X-axis
# to planets with orbits under 365 days so it's easy to read
plt.xlim(0, 365)

# Labeling our masterpiece
plt.title("Distribution of Earth-Sized Exoplanets by Orbital Period", fontsize=14, fontweight='bold')
plt.xlabel("Orbital Period (Days)", fontsize=12)
plt.ylabel("Planet Radius (Earth Radii)", fontsize=12)

# Show the plot
plt.show()
```

```python
import numpy as np

# 1. Clear out rows that don't have equilibrium temperature ('pl_eqt') data
habitable_df = clean_df.dropna(subset=['pl_rade', 'pl_eqt']).drop_duplicates(subset=['pl_name']).copy()

# 2. Define Earth's baseline values
# Earth's radius = 1.0 (in Earth radii)
# Earth's equilibrium temperature is roughly 255 Kelvin (-18°C before greenhouse effect)
R_earth = 1.0
T_earth = 255.0

# 3. Calculate a simplified Earth Similarity Index (ESI)
# ESI scales from 0 (completely different) to 1 (identical to Earth)
# We use a weighted geometric mean logic commonly used in astrobiology
weight_r = 0.57
weight_t = 1.07

esi_r = (1 - np.abs((habitable_df['pl_rade'] - R_earth) / (habitable_df['pl_rade'] + R_earth))) ** weight_r
esi_t = (1 - np.abs((habitable_df['pl_eqt'] - T_earth) / (habitable_df['pl_eqt'] + T_earth))) ** weight_t
habitable_df['ESI'] = esi_r * esi_t

# 4. Filter for top candidates (ESI > 0.75 is considered highly Earth-like)
top_candidates = habitable_df[habitable_df['ESI'] >= 0.75].sort_values(by='ESI', ascending=False)

print(f"successfully isolated {len(top_candidates)} highly Earth-like candidates based on size and temperature.")
print("\nTop 5 Most Habitable Planetary Matches:")
print(top_candidates[['pl_name', 'hostname', 'pl_rade', 'pl_eqt', 'ESI']].head(5).to_string(index=False))

# 5. Let's plot them visually!
plt.figure(figsize=(10, 6))
# Plot all planets in light gray
sns.scatterplot(data=habitable_df, x='pl_eqt', y='pl_rade', alpha=0.3, color='gray', label='All Exoplanets')
# Overlay top candidates in a bright green/blue
sns.scatterplot(data=top_candidates, x='pl_eqt', y='pl_rade', color='limegreen', s=100, edgecolor='black', label='Habitable Candidates (ESI >= 0.75)')

plt.axvline(x=T_earth, color='blue', linestyle='--', alpha=0.5, label='Earth Temp (255K)')
plt.xlim(100, 500) # Zooming into the temperate zone
plt.ylim(0, 3)     # Zooming into terrestrial sizes

plt.title("Isolating Habitable Worlds Using Earth Similarity Index (ESI)", fontsize=14, fontweight='bold')
plt.xlabel("Planet Equilibrium Temperature (Kelvin)", fontsize=12)
plt.ylabel("Planet Radius (Earth Radii)", fontsize=12)
plt.legend()
plt.show()
```
